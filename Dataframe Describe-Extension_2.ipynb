## Mehtod_2: Extension of describe-function

```python
## Absolute value
#test = data.copy()

## Simple-Returns
test = data.pct_change().dropna()

## Log-Returns
#test = np.log(data/data.shift(1)).dropna()
```

```python
stats = test.describe()
stats.loc['IQR'] = stats.loc['75%'] - stats.loc['25%']
stats = stats.append(data.reindex(stats.columns, axis=1).agg(['skew', 'mad', 'kurt']))
```

```python
def describe_extension(df):
    dfs = DataFrameSummary(df)
    dfss = stats
    return pd.concat([dfs.columns_stats.T, 
                      round(dfss.T.astype(float),3),], 
                     axis=1).T
```

```python
summary = describe_extension(test)
summary = summary.drop('count', axis = 0)
summary = summary.drop('missing', axis = 0)
summary = summary.drop('uniques', axis = 0)
summary = summary.drop('missing_perc', axis = 0)
summary = summary.drop('types', axis = 0)
```

## Mehtod_3: Table to Latex in Markdown

```python
def df_to_markdown(df, float_format='%.2g'):
    """
    Export a pandas.DataFrame to markdown-formatted text.
    DataFrame should not contain any `|` characters.
    """
    from os import linesep
    return linesep.join([
        '|'.join(df.columns),
        '|'.join(4 * '-' for i in df.columns),
        df.to_csv(sep='|', index=False, header=False, float_format=float_format)
    ]).replace('|', ' | ')
```

```python
print (df_to_markdown(df))
```
